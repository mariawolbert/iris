```python
# Import necessary modules 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 


iris = pd.read_csv("https://raw.githubusercontent.com/roholy/Iris/main/Visualization/iris.csv")
#Drop id column
iris = iris.drop('Id',axis=1)
sub=iris[iris['Species']=='Iris-setosa'] #pulls the column out of the species iris-setosa
# sns.kdeplot(data=sub[['SepalLengthCm','SepalWidthCm']], shade=True, shade_lowest=False)
x = sub['SepalLengthCm'] #calls the function sub which takes the column out of data
y = sub['SepalWidthCm']
sns.kdeplot(x,y,shade=True, shade_lowest=False)
plt.title('Iris-setosa')
plt.xlabel('Sepal Length cm')
plt.ylabel('Sepal Width cm')
plt.show()
```
